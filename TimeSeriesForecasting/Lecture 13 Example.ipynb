```python
# evaluate an ARIMA model using a walk-forward validation (Expanding Window, by default)
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt

# source of the example
# https://machinelearningmastery.com/arima-for-time-series-forecasting-with-python/
```

```python
# load dataset
def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')
series = read_csv('shampoo.csv', header=0, index_col=0, parse_dates=True, squeeze=True, date_parser=parser)
series.index = series.index.to_period('M')
series.plot()
pyplot.show()
```

```python
# split into train and test sets
X = series.values
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]                       # to list
predictions = list()
# walk-forward validation
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]                               # one step ahead forecast
    predictions.append(yhat)                       # adds one period forecast to predictions
    obs = test[t]                                  # adds one observation to history
    #history = history[1:]                         # change to Rolling window
    #print('history length=%f' % (len(history)))
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
# evaluate forecasts
rmse = sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % rmse)
```

```python
# plot forecasts against actual outcomes
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
```

```python
#now consider forecasting without rolling window
model = ARIMA(train, order=(5,1,0))
model_fit = model.fit()
output = model_fit.forecast(13)
```

```python
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.plot(output, color='green')
pyplot.show()
```

```python
# evaluate forecasts
rmse = sqrt(mean_squared_error(test, output))
print('Test RMSE: %.3f' % rmse)
```
