# Frequency Transformations

## Fourier Convolution

### Implement & Apply Fast Fourier Transform To Image (using NumPy)

```python
def fft(image, size):
    if size is not None:
        image = cv2.resize(image, size)

    fft_result = np.fft.fft2(image)
    fft_result = np.fft.fftshift(fft_result)

    return fft_result
```

```python
image_f = fft(image, None)
```

### Implement Logarithmic Frequency Image Plot Function & Plot image_f  (using NumPy, Matplotlib)

```python
def log_plot(frequency_image):
    magnitude = np.abs(frequency_image)
    log_magnitude = np.log(magnitude + 1)

    plt.imshow(((log_magnitude/np.max(log_magnitude))*255).astype(np.uint8), cmap='gray')
    plt.title('Logarithmic Frequency Image')
    plt.colorbar()
    plt.show()
```

```python
log_plot(image_f)
```
