# Sanity check DNN on the Iris dataset
--- 
confirming that there are no differences between coding styles

```python
# region imports

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score


import tensorflow as tf
import tensorflow_datasets as tfds

import keras
from keras.layers import Dense, Input
from keras.models import Sequential, Model

from sys import version

# endregion imports

version
```

```python
x, y = tfds.as_numpy(tfds.load(
    'iris',
    batch_size=-1,
    as_supervised=True,
))["train"]

x.shape, y.shape
```

```python
# converting into a single-label classification problem
(y := (y > 0) + 0)
```

```python
model = Sequential([
    Dense(8, activation='relu', input_shape=(4,)),
    Dense(1, activation='sigmoid'),
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'],
)

model.summary()
```

```python
model.fit(x, y, epochs=250)
```

```python
model.evaluate(x, y)
```

### Same model, different coding style

```python
inp = Input((4,))
est = Dense(8, activation='relu')(inp)
est = Dense(1, activation='sigmoid')(est)

model = Model(inputs=inp, outputs=est)

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'],
)

model.summary()
```

```python
model.fit(x, y, epochs=250)
```

```python
model.evaluate(x, y)
```
